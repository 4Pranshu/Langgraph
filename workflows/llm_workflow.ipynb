```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
from langchain_groq import ChatGroq
```

```python
load_dotenv()

model = ChatGroq(model="llama-3.1-8b-instant")
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def askLLM(state: LLMState) -> LLMState:
    result = model.invoke(state["question"])

    state["answer"] = str(result.content)

    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node("askLLM", askLLM)

graph.add_edge(START, "askLLM")
graph.add_edge("askLLM", END)

workflow = graph.compile()
```

```python
initial_state = {
    "question": "What is the future of AI in medical science ?",
    "answer": "",
}

final_output = workflow.invoke(initial_state)  # type:ignore
print(final_output)
```
